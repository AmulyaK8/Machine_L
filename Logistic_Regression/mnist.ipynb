```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df = pd.read_csv("/Users/amulya/Documents/GISMA Academics/Data Science/teaching-main/datasets/mnist.csv")
df
```

```python
df.set_index("id")
```

```python
#add features to X
#in the table everything except price is a feature
#axis = 1 means dropping a coloumn
#axis = 0 means dropping a row
X = df.drop(['class'],axis=1)
#add target values to y
y = df["class"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

print("df : ",df.shape)
print("X_train : ",X_train.shape)
print("X_test : ",X_test.shape)
print("y_train : ",y_train.shape)
print("y_test : ",y_test.shape)
```

```python
log_reg = LogisticRegression(max_iter=10000, random_state=0).fit(X_train, y_train)
```

Warning: if the max_iter is small, it will throw a Convergence warning. So increase the value to optimise better.

```python
y_predict = log_reg.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_predict)
accuracy
```
